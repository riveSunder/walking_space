# Reinforcement Learning and DockRL

### Reinforcement Learning is a problem formulation where an *agent* interacts with an *environment* by choosing *actions* based on the *observations* (and *rewards*) generated by the environment. 

<img src="assets/dockrl_overview.png">

Figure from my DockRL project [https://github.com/riveSunder/dockrl](https://github.com/riveSunder/dockrl). It's not _the_ DockRL project, as there are many, but a DockRL project at least.

I (partially) re-implemented the optimization experiment from Bjerrum's docking optimization paper, using covariance matrix adaptation evolution instead of particle swarm optimization and a few different parametrizations of the optimization problem (direct optimizatino of Smina objective function weights, a multilayer perceptron to generate the weights, and a naive graph neural network). 


* Bjerrum, Esben Jannik. “Machine learning optimization of cross docking accuracy.” Computational biology and chemistry 62 (2016): 133-44. [blog post](https://www.cheminformania.com/machine-learning-optimization-of-smina-cross-docking-accuracy/) [https://pubmed.ncbi.nlm.nih.gov/27179709/](https://pubmed.ncbi.nlm.nih.gov/27179709/)
