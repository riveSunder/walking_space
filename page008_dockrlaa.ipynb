# Covariance Matrix Adaptation Evolution Strategy for DockRL

### Reinforcement Learning is a problem formulation where an *agent* interacts with an *environment* by choosing *actions* based on the *observations* (and *rewards*) generated by the environment. 

<div align="center">
<img src="assets/himmelblau_es_cmaes.gif" width=80%>
</div>

Covariance Matrix Adaptation Evolution Strategy is an optimization procedure that uses the covariance matrix of the best performing agents relative to the population mean to control exploration. 

We can see in the figure above that it quickly narrows in on an optimum point in the [Himmelblau function](https://en.wikipedia.org/wiki/Himmelblau%27s_function) landscape, compared to a simple Gaussian evolution strategy with constant variance. 

* Hansen, Nikolaus. “The CMA Evolution Strategy: A Tutorial.” ArXiv abs/1604.00772 (2016):.
